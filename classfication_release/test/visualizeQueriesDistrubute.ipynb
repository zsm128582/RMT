```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

def visualize_agents_distribution(agents_tensor):
    """
    输入: agents_tensor, shape [n, dim] (例如 [64, 512])
    """
    # 1. 转换为 Numpy
    data = agents_tensor.detach().cpu().numpy()
    n_samples, dim = data.shape
    
    print(f"Analyzing {n_samples} agents with dimension {dim}...")

    # 2. PCA 可视化 (保留主成分)
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(data)
    
    # 3. t-SNE 可视化 (非线性降维)
    # perplexity 建议小于 n_samples 的 1/3
    perp = min(30, n_samples // 3) 
    tsne = TSNE(n_components=2, perplexity=perp, init='pca', learning_rate='auto', random_state=42)
    tsne_result = tsne.fit_transform(data)

    # 4. 绘图
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Plot PCA
    axes[0].scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.7, c='blue', s=50)
    axes[0].set_title(f'PCA Projection\n(Explained Var: {np.sum(pca.explained_variance_ratio_):.2f})')
    axes[0].grid(True, alpha=0.3)
    
    # Plot t-SNE
    axes[1].scatter(tsne_result[:, 0], tsne_result[:, 1], alpha=0.7, c='red', s=50)
    axes[1].set_title(f't-SNE Projection\n(Perplexity={perp})')
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


import seaborn as sns

def visualize_cosine_similarity(agents_tensor):
    # 归一化向量
    agents = agents_tensor.detach().cpu()
    agents_norm = torch.nn.functional.normalize(agents, p=2, dim=1)
    
    # 计算相似度矩阵 [n, n]
    # sim_matrix[i][j] 代表 agent_i 和 agent_j 的相似度
    sim_matrix = torch.mm(agents_norm, agents_norm.t()).numpy()
    
    # 绘图
    plt.figure(figsize=(10, 8))
    sns.heatmap(sim_matrix, cmap='viridis', vmin=-1, vmax=1)
    plt.title("Pairwise Cosine Similarity of Agents")
    plt.xlabel("Agent Index")
    plt.ylabel("Agent Index")
    plt.show()
    
    # 打印平均非对角线相似度
    mask = ~np.eye(sim_matrix.shape[0], dtype=bool)
    avg_sim = sim_matrix[mask].mean()
    print(f"平均两两相似度: {avg_sim:.4f}")

# visualize_cosine_similarity(model.agents)
# 假设你的模型 model 中有一个参数名为 agents
# visualize_agents_distribution(model.agents)
```
